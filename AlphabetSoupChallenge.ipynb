```python
# Import our dependencies
import pandas as pd
import sklearn as skl
from pandas import DataFrame
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import tensorflow as tf

# Read in our charity data
charity_df = pd.read_csv("charity_data.csv")
charity_df.dtypes
```

```python
# CREATE DATAFRAME OF NON-CATEGORICAL COLUMN, DROP EIN AND NAME COLUMNS
charity_non_encode_df = charity_df[['ASK_AMT']]
```

```python
# Create list of Categorical columns
charity_cat = ['APPLICATION_TYPE',
'AFFILIATION',
'CLASSIFICATION',
'USE_CASE',
'ORGANIZATION',
'STATUS',
'INCOME_AMT',
'SPECIAL_CONSIDERATIONS',
'IS_SUCCESSFUL']
```

```python
# Determine number of unique values per Categorical column
charity_df[charity_cat].nunique()
```

```python
# Check the unique value counts to see if binning is required
charity_df.APPLICATION_TYPE.value_counts()
```

```python
# Determine the value counts from the APPLICATION_TYPE column
app_counts = charity_df.APPLICATION_TYPE.value_counts()

# Determine which values to replace
replace_app_type = list(app_counts[app_counts < 156].index)

# Replace in DataFrame
for app in replace_app_type:
    charity_df.APPLICATION_TYPE = charity_df.APPLICATION_TYPE.replace(app,"Other")
    
# Check to make sure binning was successful
charity_df.APPLICATION_TYPE.value_counts()
```

```python
# Check the Top 10 values from the CLASSIFICATION column
charity_df.CLASSIFICATION.value_counts().nlargest(10)
```

```python
# Determine the value counts from the CLASSIFICATION column
class_counts = charity_df.CLASSIFICATION.value_counts()

# Determine which values to replace
replace_classes = list(class_counts[class_counts < 194].index)

# Replace in DataFrame
for classes in replace_classes:
    charity_df.CLASSIFICATION = charity_df.CLASSIFICATION.replace(classes,"Other")
    
# Check to make sure binning was successful
charity_df.CLASSIFICATION.value_counts()
```

```python
# Determine the consolidated value counts by column after bucketing
charity_df[charity_cat].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)
# Fit and transform the OneHotEncoder using the categorical variable list
charity_df_encoded = pd.DataFrame(enc.fit_transform(charity_df[charity_cat]))
# Add the encoded variable names to the Categorial DataFrame
charity_df_encoded.columns = enc.get_feature_names(charity_cat)
charity_df_encoded.head()
```

```python
# Added the non-classification data (ASK_AMT) to the encoded Categorical DataFrame
charity2_df = charity_non_encode_df.merge(charity_df_encoded,left_index=True,right_index=True)
charity2_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = charity2_df["IS_SUCCESSFUL_1"].values
X = charity2_df.drop(["IS_SUCCESSFUL_1", "IS_SUCCESSFUL_0"],1).values
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create the StandardScaler instance
scaler = StandardScaler()
```

```python
# Fit the StandardScaler
X_scaler = scaler.fit(X_train)
```

```python
# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  33
hidden_nodes_layer2 = 17

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

Use the tanh function in the second layer

```python
# Train the model (with tanh function in second layer)
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```
